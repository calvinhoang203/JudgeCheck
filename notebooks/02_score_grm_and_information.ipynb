# JudgeCheck Tutorial 2: 1–10 Score GRM & Test Information

**Part B** — GPT-4 scores each model 1–10. The **test information curve** shows where the benchmark best separates model quality.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))

import matplotlib.pyplot as plt
from judgecheck.pipeline import run_score_analysis

scores = run_score_analysis(ROOT / "outputs")
print("Mean discrimination:", round(scores.item_params["discrimination"].mean(), 2))
scores.information.plot(x="theta", y="test_information", figsize=(8, 4))
plt.title("Test information T(θ)")
plt.show()
```
